```python
#Basic Plotting
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('Plot of f(x) = x² - 4x + 4', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True)
plt.show()
```

```python
#. Sine and Cosine Plot
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_sin, 'b-', label='sin(x)', linewidth=2)
plt.plot(x, y_cos, 'r--', label='cos(x)', linewidth=2)
plt.title('Sine and Cosine Functions', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

```python
#Subplots
x = np.linspace(-2, 2, 100)
x_log = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 8))

# Top-left: x³
plt.subplot(2, 2, 1)
plt.plot(x, x**3, 'r-')
plt.title('f(x) = x³')
plt.xlabel('x')
plt.ylabel('f(x)')

# Top-right: sin(x)
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), 'g--')
plt.title('f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')

# Bottom-left: e^x
plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x), 'b-.')
plt.title('f(x) = e^x')
plt.xlabel('x')
plt.ylabel('f(x)')

# Bottom-right: log(x+1)
plt.subplot(2, 2, 4)
plt.plot(x_log, np.log(x_log + 1), 'm:')
plt.title('f(x) = log(x+1)')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```

```python
#Scatter Plot
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.title('Random Scatter Plot', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(True)
plt.colorbar()
plt.show()
```

```python
#Histogram
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Normal Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#3D Plotting
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('3D Surface Plot of f(x,y) = cos(x² + y²)', fontsize=14)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('f(x,y)', fontsize=12)
plt.show()
```

```python
#Bar Chart
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['blue', 'green', 'red', 'purple', 'orange']

plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=colors)
plt.title('Product Sales', fontsize=14)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales (units)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
#Stacked Bar Chart
time_periods = ['T1', 'T2', 'T3', 'T4']
categories = ['Category A', 'Category B', 'Category C']

# Sample data: rows are time periods, columns are categories
data = np.array([
    [30, 40, 30],   # T1
    [25, 45, 30],   # T2
    [20, 50, 30],   # T3
    [15, 55, 30]    # T4
])

colors = ['#FF9999', '#66B2FF', '#99FF99']

plt.figure(figsize=(8, 6))
bottom = np.zeros(len(time_periods))

for i, (category, color) in enumerate(zip(categories, colors)):
    plt.bar(time_periods, data[:, i], bottom=bottom, label=category, color=color)
    bottom += data[:, i]

plt.title('Stacked Bar Chart by Category Over Time', fontsize=14)
plt.xlabel('Time Period', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
